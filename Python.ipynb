### M치s sobre strings en Python:

* En Python 3 los strings se basan en Unicode. A principios de la computaci칩n los caracteres de un string estaban limitados a uno de 256 valores. Esto era suficiente para representar caracteres May칰sculas y min칰sculas del lat칤n, as칤 como tambi칠n d칤gitos individuales (0-9). Este lenguaje que reprensenta estos s칤mbolos es llamado ASCII y como puedes ver, era bien compacto. Sin embargo, los s칤mbolos latinos no bastan para representar muchos otros que existen de otras lenguas, o incluso s칤mbolos de otras 치reas del saber, como operadores matem치ticos.

    A ra칤z de esta necesidad es que se crea UTF  (Unicode Transformation Format), el cual es capaz de representar m치s de un mill칩n de distintos caracteres. Este no s칩lo incluye lenguajes humanos, sino que tambi칠n s칤mbolos como emojis 游땙. Python 3 utiliza UTF por defecto, as칤 no tendremos problema al representar grupos de s칤mbolos multinacionales.
    
    Junto a unicode, python utiliza un lenguaje especial para dar formato a la salida de strings. Como vimos anteriormente, con python no necesitamos explicitar el tipo de dato de una variable, sin embargo, esta ventaja tiene costos. Uno de esos costos es que si el interpretador de Python no es capaz de dar formato autom치ticamente a una variable, debemos hacer esta conversi칩n manualmente.


```python
# el interpretador no es capaz de dar formato a la concatenac칩n de un n칰mero (int) (lanza error)
print('Mi edad es : ' + 27)
```

```python
# considerando que tendr칤amos que repetir esta funci칩n c칩digos fuente de gran envergadura 
# una mejor forma de manejar estos casos ser칤a con el mini lenguaje de formateo de strings de Python:
producto = {'nombre': 'Notebook Levono r78', 'precio': 450000, 'ubicacion': 'pasillo B'}

descripcion_producto = '{} tiene un precio de {} y est치 ubicado en {}'

# da formato con funci칩n format(). Como puedes ver, el orden de los elementos influir치 en 
# el string resultante
print(descripcion_producto.format(producto['nombre'], producto['precio'], producto['ubicacion']))

# versi칩n compacta con f'strings
print(f"{producto['nombre']} tiene un precio de {producto['precio']} y est치 ubicado en {producto['ubicacion']}")
```

#### Secuencias
Estructuras comunes que generalmente creamos con loops o leyendo un archivo, las cuales permiten iterar sobre sus elementos. Python ofrece una sintaxis abreviada para su creaci칩n, llamada Listas por comprensi칩n (List Comprehension). 

```python
numeros_pares = []
for numero in range(0, 100):
    if numero % 2 == 0:
        numeros_pares.append(numero)

numeros_pares
```

* Ahora con lista por comprensi칩n, lo que resulta en un c칩digo m치s compacto y generalmente m치s r치pido. No es obligaci칩n que utilices este formato, pero s칤 debes ser capaz de reconocerlo, ya que su uso en ciencia de datos es com칰n.

```python
numeros_pares = [numero for numero in range(0,100) if numero % 2 == 0]
numeros_pares
```

쮺u치l es la diferencia con None entonces?

A pesar de su similitud, como se se침al칩, None es la forma interna/propia de Python de representar la ausencia de valor, representado con un tipo de dato Object (equivalente a null de otros lenguajes). NaN, por su parte, es un est치ndar que representa la ausencia de un n칰mero como tipo de dato float.

nan ni siquiera es igual a nan. Esto, debido a que nan, al ser un est치ndar, genera cada vez un nuevo "error" diciendo que carece de valor. 

Imaginemos por ejemplo que quisieramos representar a todas las personas que no conocemos etiquet치ndolas de "No la conozco" (NoC). As칤, cada vez que nos topemos con alguien que no conocemos, dir칤amos "No la conozco" (NoC), sin embargo, a pesar de intentar generalizar a todas las personas que no conocemos, cada una de ellas es distinta a la otra, con caracter칤sticas propias (rut, nombre, estatura, etc). En definitiva, para solucionar el problema anterior, debemos generar un nuevo tipo de error cada vez que nos topemos con alguien que no conocemos, y pese a que decimos "No la conozco" (NoC) o nan (en el caso de numpy), el error que genera es 칰nico.

#### Boolean masking

Boolean masking es el motor que da la velocidad y eficiencia a las consultas realizadas por _NumPy_ y _Pandas_. ### Valores ausentes

Previamente vimos que Pandas es capaz de manejar la falta de datos como NaN o none. Los sets de datos com칰mente poseen este tipo de valores ausentes, lo que hace a la labor de limpieza de datos parte del proceso en ciencia de datos. Otros m칠todos comunes para el manejo de datos ausente son: _na_values_, _na_filter_, entren otros. 

Tambi칠n debemos considerar los casos en los que valores faltantes representan informaci칩n. Un ejemplo para este tipo de casos es un set de datos provenientes de logs que capturan informaci칩n acerca del playback de videos educativos. Los programas enviar치n estos datos (logs) a un servidor a intervalos de tiempo especificados. Estos "latidos" de datos se ver칤an entonces de esta forma:  

```python
df = pd.read_csv('log.csv')
df
```

La primera columna es un tipo de dato timestamp (tipo de dato UNIX en formato epoch ), seguida del nombre de usuario, la p치gina web en la que se encuentran y el video que est치n visualizando. El playback position registra la posici칩n en la que se encuetran en la reproducci칩n del video, el cual va incrementando a medida avanza el tiempo a excepci칩n del usuario bob. En este caso el usuario tiene el video en pausa ya que la posici칩n no avanza en conjunto con el tiempo. N칩tese que esta informaci칩n no es f치cil de inferir debido a que los valores de la columna de timestamps no est치n en orden.

Puedes notar tambi칠n que existen muchos valores ausentes (NaN) en el data frame, lo que resulta en que no sea muy eficiente enviar estos datos a trav칠s de la red. En este caso en particular, el sistema inserta valores nulos en la base de datos si 칠stos no cambian.

```python
# asigna un nuevo index 'time' y luego reordena sus valores
df = df.set_index('time')
df = df.sort_index()
df
```

En el caso anterior, podemos observar que el index posee valores duplicados (dos o m치s usuarios pueden estar viendo un video al mismo tiempo), para manejar de mejor forma esta situaci칩n haremos un reset del index y usaremos un 칤ndice m칰ltiple.

```python
df = df.reset_index()
df = df.set_index(['time', 'user'])
df
```
